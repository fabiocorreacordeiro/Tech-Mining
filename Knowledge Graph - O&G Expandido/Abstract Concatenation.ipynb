```python
import pandas as pd
import json
from pathlib import Path
```

Lendo os arquivos JSON com os textos extraídos utilizando a API da Elsevier. Para os aquivos provenientes de artigo cujo um dos autores é afiliado a uma empresa de Óleo & Gás foi acrescentado a coluna "O&G" com o valor 1 e uma coluna "Affiliate" com o nome da empresa. Para os demais artigos foi atribuído o valor 0 para "O&G" e "outros" para a coluna "Affiliate". 

```python
#pasta de rede onde os arquivos estão localizados
source_path = '\\Users\\upe2\\Desktop\\Elsevier\\Tech Mining - Estratégia\\Knowledge Graph - O&G Expandido\\Novos Artigos O&G'
target_path = '\\Users\\upe2\\Desktop\\Elsevier\\Tech Mining - Estratégia\\Knowledge Graph - O&G Expandido'
```

```python
articles = {"Abstract": [],
            "Creator": [],
            "DOI":[],
            "Date":[],
            "Keywords":[],
            "Raw_text":[],
            "Title": []}
articles = pd.DataFrame(data=articles)
```

```python
#Iterando os arquivos da Pasta
extensao = '*.json'
pastaArquivos = Path(source_path).glob(extensao)
for path in pastaArquivos:
    articles = pd.concat([articles, 
                          pd.read_json(path_or_buf=path)])

articles
```

```python
# Carregando os artigos proveniente das empresas de O&G para preencher a coluna "Affilieate
OG = pd.read_json("Elsevier_abstract - Consolidado.json")[["Affiliate", "DOI"]]
```

```python
articles = pd.merge(articles, OG, how='left', on='DOI')
articles
```

```python
articles = articles.reset_index(drop=True)
articles.to_json("Elsevier_abstract - Consolidado.json", date_format = 'iso')
```
