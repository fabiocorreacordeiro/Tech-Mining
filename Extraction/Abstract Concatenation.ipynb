Importando bibliotecas

```python
import pandas as pd
import json
```

O objetivo deste notebook é concatenar os arquivos JSON com informações sobre os artigos extraídos utilizando a API da Elsevier. Foi acrescentada a coluna "Affiliate" com o nome da empresa.

```python
# Inicialmete lemos o arquivo referente aos textos da "Petrobras"
abstract_petrobras = pd.read_json(path_or_buf="Elsevier_abstracts - Petrobras.json")
abstract_petrobras['Affiliate'] = 'Petrobras'
abstract_petrobras
```

```python
# Lendo os arquivos das demais Majors
abstract = abstract_petrobras
#Lista com o nome das demais majors
companies = ["BP America",
             "BP Exploration",
             "BP International",
             "BP United Kingdom",
             "British Petroleum",
             "Chevron",
             "Eni",
             "Equinor",
             "Exxon",
             "Repsol",
             "Shell",
             "Statoil",
             "Total"]

# Iterando sobre todas as majors
for company in companies:
    #Lendo o arquivo de cada compania
    path = "Elsevier_abstracts - " + company + ".json"
    abstract_company = pd.read_json(path_or_buf = path)
    abstract_company['Affiliate'] = company
    #Concatenando com o dataframe consolidado
    abstract = pd.concat([abstract, abstract_company])
    
abstract
```

Salvando arquivo consolidado

```python
abstract = abstract.reset_index(drop=True)
abstract.to_json("Elsevier_abstract - Consolidado.json", date_format = 'iso')
```
