### Base de teste de Creator

A base de teste será composta por três colunas. As duas primeiras são de autores que publicaram em conjunto em um mesmo artigo, a terceira coluna conterá uma autor aleatório. Espera-se que o vetor do autor da primeira coluna esteja mais próximo do autor da segunda coluna do que da terceira. Para compor o corpus de teste de autores apenas serão considerados os 5 primeiros autores de cada artigo.

```python
max_creator = 5                             #Número máximo de autores que será considerado
n_article = len(abstracts)
creator_list = creator["Creator"].tolist() # Lista de Autores

# Criando um DataFrame para receber os pares de Autores que aparecem no mesmo artigo
# e o autor aleatório
test_corpus_creator = pd.DataFrame(columns=["Creator_1", "Creator_2", "Random_Creator"])

for article in abstracts.iterrows():  # Iterando sobre todos os artigos
    
    words = article[1]["Creator"]     # Célula do Dataframe onde se encontram os autores
    if words == []:                   # Se não houver keyword passa para próximo artigo
        pass
    
    else:
        for n in range(len(words)):   # Iterando sobre todos os autores de um artigo
            if n >= max_creator:      # Parar de iterar após o número máximo de autores
                break
            
            word_1 = words[n]         # Identificando primeiro autor
            
            # Como nem todos os autores estão na Lista de Autores é necessário 
            # verificar se word_1 está em creator_list 
            
            if word_1 in creator_list:
                for m in range(n+1, len(words)): # Iterando sobre os demais autores
                    if m >= max_creator:         # Parar de iterar após o número máximo de autores
                        break
                    
                    word_2 = words[m]            # Identificando segund0 autor
                    if word_2 in creator_list:
                        # Identificando autor aleatório
                        random_word = random.sample(creator_list, 1)[0]
                        # Adicionando os três autores no dataframe "test_corpus_creator"
                        test_corpus_creator.loc[len(test_corpus_creator)]= [word_1, word_2, random_word]
```

```python
test_corpus_creator.to_csv("Test_Corpus_creator.csv")
```
